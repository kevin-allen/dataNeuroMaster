## Scatter plots to look for violation of the pearson correlation assumptions

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
nSize=50
fig, ax = plt.subplots(2,2,figsize=(10,7))

a = np.random.normal(loc=0,scale=5,size=nSize)
b = a**2 + 20 + np.random.normal(loc=10,scale=15,size=nSize)
ax[0,0].scatter(a,b)

a = np.random.normal(loc=0,scale=5,size=nSize)
b = np.random.poisson(lam=1.5, size=nSize)
ax[1,0].scatter(a,b)


a = np.random.normal(loc=0,scale=5,size=nSize)
b = np.random.normal(loc=0,scale=5,size=nSize)
a=np.append(a,25)
b=np.append(b,25)
ax[0,1].scatter(a,b)


nSize=500
a = np.random.normal(loc=0,scale=5,size=nSize)
b = stats.skewnorm.rvs(a=100,size=nSize)+2
ax[1,1].scatter(a,b)
ax[1,1].set_ylim(1,6)

fn = "../images/pearsonAssumptionViolations.png"
plt.savefig(fn)
```
